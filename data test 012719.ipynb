```python
#Maybe a good example of Data cleaning/"munging" in python?
import pandas as pd
import numpy as np
```

```python
co_data = pd.read_csv('history_co2.csv',skiprows=3) #read the csv, skip the first 3 rows
#hourly data from CO sensor, 9 months of 15-min-interval data from 01/2018 to 09/2018
```

```python
co_data = co_data.drop(['Trend Flags','Status'], axis=1)   #delete unnecessary columns
```

```python
co_data.head()
```

```python
co_data['Timestamp']=pd.to_datetime(co_data['Timestamp'])  
co_data['Timestamp_copy'] = co_data['Timestamp']
co_data = co_data.set_index(['Timestamp'])
#change format of timestamp to what python can recognize, set index as timestamp
#make a copy of timestamp col for reference later
```

```python
co_data.head()
```

```python
co_data['CO_values']=co_data['Value (ppm)'].str[0:4]    
#pull the first 4 characters of the Value(ppm) column in order to convert them to numeric, below
co_data.head()
```

```python
co_data.dtypes
```

```python
co_data['CO_values']=pd.to_numeric(co_data['CO_values']) #convert from strings to number
```

```python
co_data.dtypes
```

```python
Non_zero = co_data[co_data['CO_values']!=0]   #Filtering to see non-zero values if needed
Non_zero.head()
```

```python
co_data['Weekday?'] = co_data['Timestamp_copy'].dt.dayofweek  
#add a column that indicate weekday for logic later.. (0=Mon, 6=Sun)
```

```python
co_data['weekday_TF'] = (co_data['Weekday?']<5)  
#if weekday # is less than 6, it's not 5 or 6(sat or sun)
#add a boolean column
```
